```python
%matplotlib inline
from sklearn import datasets
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
import numpy as np
```

```python
iris = datasets.load_iris()
```

```python
X = iris.data[:, :2]
y = iris.target
```

```python
plt.scatter(X[:,0], X[:,1], c=y, cmap='gist_rainbow')
plt.xlabel('Spea1 Length', fontsize=18)
plt.ylabel('Sepal Width', fontsize=18)
```

```python
km = KMeans(n_clusters = 4, n_jobs = 4, random_state=21)
km.fit(X)
```

```python
centers = km.cluster_centers_
print(centers)
```

```python
print(km.inertia_)
```

```python
new_labels = km.labels_
fig, axes = plt.subplots(1, 2, figsize=(16,8))
axes[0].scatter(X[:, 0], X[:, 1], c=y, cmap='gist_rainbow',edgecolor='k', s=150)
axes[1].scatter(X[:, 0], X[:, 1], c=new_labels, cmap='jet',edgecolor='k', s=150)
axes[0].set_xlabel('Sepal length', fontsize=18)
axes[0].set_ylabel('Sepal width', fontsize=18)
axes[1].set_xlabel('Sepal length', fontsize=18)
axes[1].set_ylabel('Sepal width', fontsize=18)
axes[0].tick_params(direction='in', length=10, width=5, colors='k', labelsize=20)
axes[1].tick_params(direction='in', length=10, width=5, colors='k', labelsize=20)
axes[0].set_title('Actual', fontsize=18)
axes[1].set_title('Predicted', fontsize=18)
```

```python
samples = np.array([
    [4., 4.],
    [6., 2.5],
    [7., 3.],
])

scaler = StandardScaler()
kmeans = KMeans(n_clusters=3)

pipeline = make_pipeline(scaler, kmeans)
```

```python
pipeline.fit(samples)
labels = pipeline.predict(samples)

fig, axes = plt.subplots(1, 1, figsize=(16,8))
axes.scatter(samples[:, 0], samples[:, 1], c=labels, cmap='jet',edgecolor='k', s=150)
axes.set_xlabel('Sepal length', fontsize=18)
axes.set_ylabel('Sepal width', fontsize=18)
axes.tick_params(direction='in', length=10, width=5, colors='k', labelsize=20)
axes.set_title('Predicted', fontsize=18)
```
